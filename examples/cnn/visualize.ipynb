```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import jax.numpy as jnp
import jax
import seaborn as sns
from jax.numpy import float64
jax.config.update("jax_enable_x64", True)
run_csv = pd.read_csv('loss_vs_iter.csv')
```

```python
# Visualze Training Loss
# Remove the columns ends with _MAX or _MIN and remove Step column
run_csv = run_csv.loc[:,~run_csv.columns.str.endswith('_MAX')]
run_csv = run_csv.loc[:,~run_csv.columns.str.endswith('_MIN')]
run_csv = run_csv.loc[:,~run_csv.columns.str.endswith('Step')]

# X-axis: Steps
# Y-axis: Mean-Squared Error
# Title: Edge Detection
# Line: Mean and shaded area for the min/max

# Get the mean
mean = run_csv.mean(axis=1)

# Plot the mean and min/max values
steps = np.arange(0, len(mean))

# Plotting
plt.figure(figsize=(5, 4))  # Set a larger figure size
plt.plot(steps, mean, color='blue', linewidth=3)  # Thicker line

# Fill between min and max values
plt.fill_between(steps, run_csv.min(axis=1), run_csv.max(axis=1), color='blue', alpha=0.3)

# Enhancing aesthetics
# plt.title('Edge Detection', fontsize=18, fontweight='bold')  # Even larger, bold title
plt.xlabel('Steps', fontsize=14)  # Larger label font size
plt.ylabel('Mean-Squared Error', fontsize=14)  # Larger label font size

# Adjust tick size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add a grid with a finer style
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)

# Set y-axis limits for better focus
plt.ylim(0.1, 0.7)

# Modify legend
# plt.legend(loc='upper right', fontsize=12, frameon=True, shadow=True)

# Adjust layout to prevent cutting off labels
plt.tight_layout()

# Show the plot
plt.savefig('cnn_loss_vs_iter.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
init_weight = jnp.array([ 0. ,  0. ,  0. ,  0. ,  2. ,  0. ,  0. ,  0. ,  0. , -1. , -1. ,-1. , -1. ,  8. , -1. , -1. , -1. , -1. , -0.5])

def plot_heatmap(matrix: jnp.ndarray, ax: plt.Axes, vmin: float, vmax: float, cmap='coolwarm'):

    sns.heatmap(matrix, ax=ax, cmap=cmap, cbar=False, 
                vmin=vmin, vmax=vmax, annot=True, square=True, linewidths=1, linecolor='black',annot_kws={"size": 16})

    ax.set_xticks([])
    ax.set_yticks([])

def mat_to_ABZ(w: jnp.ndarray):
    # weight consisits of A (flattened 3x3) + B (flattened 3x3) + z (1-d)
    # Plot each component as a heatmap on an axis
    A_mat = w[:9].reshape(3, 3)
    B_mat = w[9:-1].reshape(3, 3)
    z = w[-1]
    return A_mat, B_mat, z


# Plot all weights
# 3 rows, len(weights) + 1 columns
# Rows visualize A, B, z respectively
# The fisrt column is the initial weights
# All heatmaps share the same colorbar
fig, axs = plt.subplots(3, len(weights) + 1, figsize=(26, 8))
vmin = min([w[0].min() for w in weights])
vmax = max([w[0].max() for w in weights])

# Row labels
row_labels = ["A", "B", "z"]

init_A, init_B, init_z = mat_to_ABZ(init_weight)
plot_heatmap(init_A, axs[0, 0], vmin, vmax)
plot_heatmap(init_B, axs[1, 0], vmin, vmax)
plot_heatmap(jnp.array([[init_z]]), axs[2, 0], vmin, vmax)


for i, w in enumerate(weights):
    A, B, z = mat_to_ABZ(w[0])
    plot_heatmap(A, axs[0, i + 1], vmin, vmax)
    plot_heatmap(B, axs[1, i + 1], vmin, vmax)
    plot_heatmap(jnp.array([[z]]), axs[2, i + 1], vmin, vmax)

# Add row titles (A, B, z)
for row_idx, label in enumerate(row_labels):
    axs[row_idx, 0].set_ylabel(label, fontsize=35, fontweight='bold', rotation=0, labelpad=20, va='center')


# Adjust subplot spacing
plt.tight_layout()
plt.subplots_adjust(wspace=0.1, hspace=0.1)  # Adjust spacing between subplots

# Place the colorbar on rightmost side
cbar_ax = fig.add_axes([1.0, 0.15, 0.02, 0.7])
sm = plt.cm.ScalarMappable(cmap='coolwarm', norm=plt.Normalize(vmin=vmin, vmax=vmax))
cbar = fig.colorbar(sm, cax=cbar_ax)
cbar.ax.tick_params(labelsize=16)

plt.savefig('cnn_weights.pdf', dpi=300, bbox_inches='tight')
plt.show()

# Plot the average weight difference
# Calculate the average weight difference
A_diffs, B_diffs, z_diffs = [], [], []
for i, w in enumerate(weights):
    A, B, z = mat_to_ABZ(w[0])
    A_diffs.append(A - init_A)
    B_diffs.append(B - init_B)
    z_diffs.append(z - init_z)

A_diffs = jnp.mean(jnp.stack(A_diffs), axis=0)
B_diffs = jnp.mean(jnp.stack(B_diffs), axis=0)
z_diffs = jnp.mean(jnp.stack(z_diffs), axis=0)

fig, axs = plt.subplots(3, 1, figsize=(4, 8))
vmin = min([A_diffs.min(), B_diffs.min(), z_diffs.min()])
vmax = max([A_diffs.max(), B_diffs.max(), z_diffs.max()])
# make the colorbar symmetric
vmax = max(abs(vmin), abs(vmax))
vmin = -vmax

plot_heatmap(A_diffs, axs[0], vmin, vmax, cmap="BrBG")
plot_heatmap(B_diffs, axs[1], vmin, vmax, cmap="BrBG")
plot_heatmap(jnp.array([[z_diffs]]), axs[2], vmin, vmax, cmap="BrBG")

plt.tight_layout()
# Place the colorbar on rightmost side
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
sm = plt.cm.ScalarMappable(cmap="BrBG", norm=plt.Normalize(vmin=vmin, vmax=vmax))
cbar = fig.colorbar(sm, cax=cbar_ax)
cbar.ax.tick_params(labelsize=16)

plt.savefig("cnn_weights_diff.pdf", dpi=300, bbox_inches="tight")
plt.show()

# Plot the initial weights
# Calculate the average weight difference

fig, axs = plt.subplots(3, 1, figsize=(4, 8))
vmin = min([init_A.min(), init_B.min(), init_z.min()])
vmax = max([init_A.max(), init_B.max(), init_z.max()])

plot_heatmap(init_A, axs[0], vmin, vmax)
plot_heatmap(init_B, axs[1], vmin, vmax)
plot_heatmap(jnp.array([[init_z]]), axs[2], vmin, vmax)
# Add row titles (A, B, z)
for row_idx, label in enumerate(row_labels):
    axs[row_idx].set_ylabel(label, fontsize=35, fontweight='bold', rotation=0, labelpad=20, va='center')

plt.tight_layout()
# Place the colorbar on rightmost side
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
sm = plt.cm.ScalarMappable(cmap='coolwarm', norm=plt.Normalize(vmin=vmin, vmax=vmax))
cbar = fig.colorbar(sm, cax=cbar_ax)
cbar.ax.tick_params(labelsize=16)

plt.savefig('cnn_init_weights.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
